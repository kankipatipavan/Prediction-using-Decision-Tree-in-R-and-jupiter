# Prediction using Decision Tree Algorithm

## Import Libraries

```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas.util.testing as tm
```

### Load data :

```python
from sklearn.datasets import load_iris
dat=load_iris()
data=pd.DataFrame(dat.data, columns=dat.feature_names)
y=dat.target
data.head(8)
```

# Pair plots :

```python
sns.pairplot(data)
```

## Training and Test Data

```python
x=data
y=y
from sklearn .model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state = 42)

print("Shape of feature training data:",x_train.shape)
print("Shape of target training data:",y_train.shape)
print("Shape of feature test data:",x_test.shape)
print("Shape of target test data:",y_test.shape)
```

## Decission Tree classifier :

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score,confusion_matrix

clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
clf
```

### Predicting on train data 

```python
predict = clf.predict(x_train)
print("Accuracy of training data :",accuracy_score(predict,y_train)*100,"%")
print("Confusion matrix of training data:'\n'",confusion_matrix(predict,y_train))
sns.heatmap(confusion_matrix(predict,y_train),annot = True,cmap = 'RdYlGn')
```

### Predicting on test data

```python
predict = clf.predict(x_test)
print("Accuracy of testing data :",accuracy_score(predict,y_test)*100,"%")
print("Confusion matrix of testing data:'\n'",confusion_matrix(predict,y_test))
sns.heatmap(confusion_matrix(predict,y_test),annot = True,cmap = 'RdYlGn')
```

## Decission tree 

```python
plt.figure(figsize = (20,20))
tree.plot_tree(clf,filled = True,rounded = True,proportion = True, node_ids = True , feature_names = dat.feature_names)
plt.show()
```
